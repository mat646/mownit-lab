```julia
f(x) = 1/(1 + 25*x^2)
```

```julia
# interpolacja lagrangea

y1 = []

for i in -1:0.1:2
    push!(y1, f(i))
end
```

```julia
function lagrange(f::Function, ran, points)
    y2 = []
    for x in ran
        r = 0
        for i in points
            a = 1
            for j in points
                if (j != i)
                    a *= (x - j) / (i - j)
                end
            end
            r += a*f(i)
        end
        push!(y2, r)
    end
    return y2
end
```

```julia

lagr = lagrange(f, -1:0.1:2, -1:1)

x = -1:0.1:2; y = hcat(y1, lagr); # These are the plotting data
plot(x,y)
```

```julia
# węzły czebyszewa

x = 0.1

t0 = 1
t1 = x

t = [t0, t1]

for i in 1:10
    
    tk = 2*x*t1 - t0
    
    push!(t, tk)
    
    t0 = t1
    t1 = tk
    
end
```

```julia
t = sort(t)
```

```julia
lagr5 = lagrange(f, -1:0.01:1, t)

y5 = []

for i in -1:0.01:1
    push!(y5, f(i))
end

x = -1:0.01:1; y = hcat(y5, lagr5);

plot(x,y)
```
